```python
import pickle
from os import path
import jieba
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import sys
import time
import numpy as np
import PIL.Image as Image

reload(sys)
sys.setdefaultencoding('utf8')

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] =False
```

```python
text = open('summary.txt').read()
```

```python
wordlist = jieba.cut(text, cut_all = True)
word_space_split = " ".join(wordlist)
#background_Image = np.array(Image.open("meerca_2.jpg"))
background_Image = plt.imread("meerca_3.jpg")
wc = WordCloud(background_color="white",max_words=2000,
                mask=background_Image, max_font_size=60,random_state=32,scale=8,
                font_path="C:\Windows\Fonts\msyh.ttc").generate(word_space_split)
```

```python
#改变字体颜色
img_colors = ImageColorGenerator(background_Image)
#字体颜色为背景图片的颜色
wc.recolor(color_func=img_colors)
plt.imshow(wc)
plt.axis("off")#不显示x轴y轴
plt.savefig("summary{}.jpg".format(time.time()),dpi=1000) #用反斜杠的话会报错
plt.show()
```
